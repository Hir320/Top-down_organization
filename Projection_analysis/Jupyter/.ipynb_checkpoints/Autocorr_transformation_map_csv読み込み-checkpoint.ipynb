```python
import pandas as pd
import matplotlib.pyplot as plt
from skimage.measure import find_contours
import numpy as np
import tifffile
from pathlib import Path

# --- Parameters ---
roi_size = 330  # Must match the ROI size used during metric extraction

# --- Define file paths for the transformed map ---
# In this example, the transformed image and its metrics are stored with a "_transformed" suffix,
# and the transformed image is registered to match the target mask (e.g. Mask_118.tif).
base_name = "LGN_99"
csv_file_path = Path("metrics_df") / f"{base_name}_transformed.csv"
tif_file_path = Path(f"{base_name}.tif")
mask_file_path = Path("Mask_118.tif")  # e.g. 'Mask_99.tif'

# --- Read the metrics CSV ---
metrics_df = pd.read_csv(csv_file_path)

# --- Extract pixel size from the transformed TIFF file metadata ---
with tifffile.TiffFile(tif_file_path) as tif:
    page = tif.pages[0]
    x_res_fraction = page.tags['XResolution'].value
    pixel_size = x_res_fraction[0] / x_res_fraction[1]
    print("Pixel size:", pixel_size, "micron/pixel")

# --- Adjust orientation values to the range [0, 180) ---
metrics_df['orientation_max_degrees'] = metrics_df['orientation_max_degrees'] % 180
metrics_df['orientation_max_radians'] = np.deg2rad(metrics_df['orientation_max_degrees'])

# --- Compute ROI center coordinates in microns ---
roi_center_x = (metrics_df['x'] + roi_size/2) * pixel_size
roi_center_y = (metrics_df['y'] + roi_size/2) * pixel_size

# --- Load the transformed mask and compute its contours ---
mask_img = tifffile.imread(mask_file_path)
mask_img = mask_img.astype(float)
mask_binary = mask_img > 0.5

# Convert mask contours from pixels to microns.
contours = find_contours(mask_binary, level=0.5)
contours_microns = [contour * pixel_size for contour in contours]

# --- Plotting ---
# Set fixed ranges for the color maps (adjust these as needed)
vmin_rhalf, vmax_rhalf = 0, 15           # for r_half_microns
vmin_neg, vmax_neg = 0, 0.15             # for positive-converted first_negative_amp
vmin_second, vmax_second = 30, 150       # for second_positive_distance_microns

# Create a 2x2 subplot layout.
fig, axs = plt.subplots(2, 2, figsize=(18, 12))
axs = axs.ravel()  # flatten to 1D array for easier indexing

# Panel 1: r_half_microns (50% width)
sc1 = axs[0].scatter(roi_center_x, roi_center_y, 
                     c=metrics_df['r_half_microns'], cmap='viridis', s=20,
                     vmin=vmin_rhalf, vmax=vmax_rhalf)
for contour in contours_microns:
    axs[0].plot(contour[:, 1], contour[:, 0], linewidth=2, color='white')
axs[0].set_title("r_half_microns (50% width)")
axs[0].set_xlabel("X (microns)")
axs[0].set_ylabel("Y (microns)")
axs[0].set_aspect('equal', adjustable='box')
cbar1 = fig.colorbar(sc1, ax=axs[0])
cbar1.set_label("r_half (microns)")

# Panel 2: first_negative_amp (converted to positive)
first_negative_amp_positive = -metrics_df['first_negative_amp']
sc2 = axs[1].scatter(roi_center_x, roi_center_y, 
                     c=first_negative_amp_positive, cmap='plasma', s=20,
                     vmin=vmin_neg, vmax=vmax_neg)
for contour in contours_microns:
    axs[1].plot(contour[:, 1], contour[:, 0], linewidth=2, color='white')
axs[1].set_title("First Negative Peak Amplitude (Positive)")
axs[1].set_xlabel("X (microns)")
axs[1].set_ylabel("Y (microns)")
axs[1].set_aspect('equal', adjustable='box')
cbar2 = fig.colorbar(sc2, ax=axs[1])
cbar2.set_label("Amplitude (abs value)")

# Panel 3: second_positive_distance_microns
sc3 = axs[2].scatter(roi_center_x, roi_center_y, 
                     c=metrics_df['second_positive_distance_microns'], cmap='inferno', s=20,
                     vmin=vmin_second, vmax=vmax_second)
for contour in contours_microns:
    axs[2].plot(contour[:, 1], contour[:, 0], linewidth=2, color='white')
axs[2].set_title("Second Positive Peak Distance")
axs[2].set_xlabel("X (microns)")
axs[2].set_ylabel("Y (microns)")
axs[2].set_aspect('equal', adjustable='box')
cbar3 = fig.colorbar(sc3, ax=axs[2])
cbar3.set_label("Distance (microns)")

# Panel 4: Orientation map as tilted lines.
line_length = 20  # in microns; adjust as needed.
angles_rad = metrics_df['orientation_max_radians']  # already adjusted to [0, )
x_end = roi_center_x + line_length * np.cos(angles_rad)
y_end = roi_center_y + line_length * np.sin(angles_rad)
for xc, yc, xe, ye in zip(roi_center_x, roi_center_y, x_end, y_end):
    axs[3].plot([xc, xe], [yc, ye], color='blue', linewidth=2)
# Plot ROI centers as small black dots.
axs[3].scatter(roi_center_x, roi_center_y, color='black', s=10)
for contour in contours_microns:
    axs[3].plot(contour[:, 1], contour[:, 0], linewidth=2, color='white')
axs[3].set_title("Orientation Map (Tilted Lines, 0-180)")
axs[3].set_xlabel("X (microns)")
axs[3].set_ylabel("Y (microns)")
axs[3].set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.interpolate import griddata
import tifffile
from pathlib import Path

# --- Parameters ---
roi_size = 330  # Must match the ROI size used during metric extraction
# Assume pixel_size is already determined (e.g., from the TIFF metadata)
pixel_size = 0.640221  # micron/pixel

# --- Load metrics_df ---
base_name = "RSC_80"
csv_file_path = Path("metrics_df") / f"{base_name}_transformed.csv"
metrics_df = pd.read_csv(csv_file_path)

# --- Compute ROI center coordinates in microns ---
roi_center_x = (metrics_df['x'] + roi_size/2) * pixel_size
roi_center_y = (metrics_df['y'] + roi_size/2) * pixel_size

# --- Define grid for interpolation ---
# Determine the extent from the ROI centers
min_x, max_x = np.min(roi_center_x), np.max(roi_center_x)
min_y, max_y = np.min(roi_center_y), np.max(roi_center_y)

# Create a grid with 1 micron spacing (you can adjust as needed)
num_x = int(np.round(max_x - min_x)) + 1
num_y = int(np.round(max_y - min_y)) + 1
xi = np.linspace(min_x, max_x, num_x)
yi = np.linspace(min_y, max_y, num_y)
XI, YI = np.meshgrid(xi, yi)

# --- Interpolate metrics onto the grid ---
# Metric 1: r_half_microns
Z1 = griddata((roi_center_x, roi_center_y), metrics_df['r_half_microns'], (XI, YI), method='cubic')
Z1_m = np.ma.masked_invalid(Z1).filled(0)

# Metric 2: first_negative_amp (converted to positive)
first_negative_amp_positive = -metrics_df['first_negative_amp']
Z2 = griddata((roi_center_x, roi_center_y), first_negative_amp_positive, (XI, YI), method='cubic')
Z2_m = np.ma.masked_invalid(Z2).filled(0)

# Metric 3: second_positive_distance_microns
Z3 = griddata((roi_center_x, roi_center_y), metrics_df['second_positive_distance_microns'], (XI, YI), method='cubic')
Z3_m = np.ma.masked_invalid(Z3).filled(0)

# Metric 4: orientation_max_degrees (or you could choose orientation_max_radians)
Z4 = griddata((roi_center_x, roi_center_y), metrics_df['orientation_max_degrees'], (XI, YI), method='cubic')
Z4_m = np.ma.masked_invalid(Z4).filled(0)

# --- Save the interpolated arrays as grayscale TIFF images ---
output_dir = Path("metrics_maps")
output_dir.mkdir(exist_ok=True)

tifffile.imwrite(output_dir / f"{base_name}_r_half_microns.tif", Z1_m.astype(np.float32))
tifffile.imwrite(output_dir / f"{base_name}_first_negative_amp_positive.tif", Z2_m.astype(np.float32))
tifffile.imwrite(output_dir / f"{base_name}_second_positive_distance_microns.tif", Z3_m.astype(np.float32))
tifffile.imwrite(output_dir / f"{base_name}_orientation_max_degrees.tif", Z4_m.astype(np.float32))

print("Four TIFF images have been saved in the 'metrics_maps' directory:")
print(" - r_half_microns.tif")
print(" - first_negative_amp_positive.tif")
print(" - second_positive_distance_microns.tif")
print(" - orientation_max_degrees.tif")
```

```python
%%time
import numpy as np
import pandas as pd
from scipy.interpolate import griddata
import tifffile
from pathlib import Path
import multiprocessing as mp

# --- Parameters ---
roi_size = 330  # Must match the ROI size used during metric extraction
pixel_size = 0.640221  # micron/pixel (example value)

# --- Load metrics_df ---
base_name = "ACC_82"
csv_file_path = Path("metrics_df") / f"{base_name}_transformed.csv"
metrics_df = pd.read_csv(csv_file_path)

# --- Compute ROI center coordinates in microns ---
roi_center_x = (metrics_df['x'] + roi_size/2) * pixel_size
roi_center_y = (metrics_df['y'] + roi_size/2) * pixel_size

# --- Define grid for interpolation ---
# Determine grid extent from ROI centers
min_x, max_x = np.min(roi_center_x), np.max(roi_center_x)
min_y, max_y = np.min(roi_center_y), np.max(roi_center_y)
num_x = int(np.round(max_x - min_x)) + 1
num_y = int(np.round(max_y - min_y)) + 1
xi = np.linspace(min_x, max_x, num_x)
yi = np.linspace(min_y, max_y, num_y)
XI, YI = np.meshgrid(xi, yi)

# Prepare points array for interpolation
points = np.column_stack((roi_center_x, roi_center_y))

# --- Define a helper function for parallel interpolation ---
def interpolate_metric_chunk(args):
    """
    Interpolates a chunk of the grid.
    
    Parameters:
      args: tuple containing (values, y_range, XI, YI, method)
    Returns:
      Interpolated grid chunk.
    """
    values, y_range, XI, YI, method = args
    y_start, y_end = y_range
    XI_chunk = XI[y_start:y_end, :]
    YI_chunk = YI[y_start:y_end, :]
    Z_chunk = griddata(points, values, (XI_chunk, YI_chunk), method=method)
    return Z_chunk

def parallel_interpolation(values, XI, YI, method='linear'):
    num_cores = mp.cpu_count()
    n_rows = XI.shape[0]
    chunk_size = n_rows // num_cores
    # Create row ranges for each chunk
    ranges = [(i * chunk_size, n_rows if i == num_cores - 1 else (i + 1) * chunk_size)
              for i in range(num_cores)]
    args_list = [(values, r, XI, YI, method) for r in ranges]
    with mp.Pool(num_cores) as pool:
        results = pool.map(interpolate_metric_chunk, args_list)
    Z = np.vstack(results)
    return Z

# --- Interpolate metrics in parallel ---

# Metric 1: r_half_microns
values1 = metrics_df['r_half_microns'].values
Z1 = parallel_interpolation(values1, XI, YI, method='cubic')
Z1_m = np.ma.masked_invalid(Z1).filled(0)

# Metric 2: first_negative_amp (converted to positive)
first_negative_amp_positive = -metrics_df['first_negative_amp'].values
Z2 = parallel_interpolation(first_negative_amp_positive, XI, YI, method='cubic')
Z2_m = np.ma.masked_invalid(Z2).filled(0)

# Metric 3: second_positive_distance_microns
values3 = metrics_df['second_positive_distance_microns'].values

#  1) dedupe (remove exact duplicate (x,y) points) 
coords, idx    = np.unique(points, axis=0, return_index=True)
uniq_vals3     = values3[idx]
Z3_unique      = griddata(coords, uniq_vals3, (XI, YI), method='cubic')
n_finite_unique = np.count_nonzero(np.isfinite(Z3_unique))
print("uniquepts cubic finite:", n_finite_unique, "/", Z3_unique.size)

#  2) choose which result to use 
if n_finite_unique > 0:
    # cubic on unique points succeeded
    Z3 = Z3_unique
    coverage = n_finite_unique / Z3_unique.size
    print(f"cubic coverage: {coverage:.1%}")
else:
    # cubic failed even on unique points  try a different method
    # you could switch to linear or nearest, or your parallel_interpolation
    print("cubic failed; falling back to linear interpolation")
    Z3 = parallel_interpolation(values3, XI, YI, method='linear')

#  3) mask and fill for saving 
Z3_m = np.ma.masked_invalid(Z3).filled(0)

# Metric 4: orientation_max_degrees (or orientation_max_radians)
values4 = metrics_df['orientation_max_degrees'].values
Z4 = parallel_interpolation(values4, XI, YI, method='cubic')
Z4_m = np.ma.masked_invalid(Z4).filled(0)

# --- Save the interpolated arrays as grayscale TIFF images ---
output_dir = Path("metrics_maps")
output_dir.mkdir(exist_ok=True)

tifffile.imwrite(output_dir / f"{base_name}_r_half_microns.tif", Z1_m.astype(np.float32))
tifffile.imwrite(output_dir / f"{base_name}_first_negative_amp_positive.tif", Z2_m.astype(np.float32))
tifffile.imwrite(output_dir / f"{base_name}_second_positive_distance_microns.tif", Z3_m.astype(np.float32))
tifffile.imwrite(output_dir / f"{base_name}_orientation_max_degrees.tif", Z4_m.astype(np.float32))

print("Four TIFF images have been saved in the 'metrics_maps' directory:")
print(f" - {base_name}_r_half_microns.tif")
print(f" - {base_name}_first_negative_amp_positive.tif")
print(f" - {base_name}_second_positive_distance_microns.tif")
print(f" - {base_name}_orientation_max_degrees.tif")
```
